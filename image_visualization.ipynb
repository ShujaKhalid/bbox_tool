```python
import pandas as pd
import numpy as np
import glob
```

```python
PATH_TO_TOOLS_FILE = './../csv/'
PATH_TO_VIDS = './../ErrorVids/'
OSATS_FILE = './../OSATS/Period 2 and 3 Surgeon Roles.xlsx'
OUTPUT_FOLDER_CLIPS = './../tool_clips/'
OUTPUT_FOLDER_FRAMES = './../tool_frames/'
TOOLS = ['Veress Needle', 'Trocar', 'Kelly Forceps', 'Nathanson Retractor',
 'Marking Needle', 'Bowel Grasper', 'Ultrasonic Dissector', 'Stapler',
 'Clip Applier', 'Clip ', 'Needle Driver', 'Scissors', 'Suture Needle',
 'Maryland', 'Suction', 'Suture', 'Bougie', 'Endo-Close', 'Retractor', 'L-hook',
 'Camera / Image', 'Advanced Bipolar',  'Endo-Bag', 'Monopolar']
```
